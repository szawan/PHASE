```python
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
phase_data = [0.83, 0.89, 0.82, 0.86, 0.83]
linear_data = [0.51, 0.58, 0.67, 0.59, 0.47]

data_all = [phase_data, linear_data]

kw_statistic, kw_p_value = stats.kruskal(phase_data, linear_data)

print("Kruskal-Wallis H-test Statistic:", kw_statistic)
print("p-value:", kw_p_value)

plt.figure(figsize=(5, 8))
box = plt.boxplot(data_all, labels=['PHASE', 'Linear'], patch_artist=True, widths=0.6)

colors = ['lightblue', "#4682B4"]
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

for i in range(len(phase_data)):
    plt.plot([1, 2], [phase_data[i], linear_data[i]], color='gray', linestyle='--', marker='o', markersize=5)


plt.ylabel('Correlation')
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig('./Compare/performance_phase_linear.pdf')
plt.show()
```
